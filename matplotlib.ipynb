# **Matplotlib Tutorial**

Matplotlib is a low level graph plotting library in python that serves as a visualiztion utility.

Matplotlib was created by John D. Hunter.

It is an open source and free.

**Matplotlib Source Code:** https://github.com/matplotlib/matplotlib

**Matplotlib Getting Started**

---



```python
# check the version of matplotlib

import matplotlib
print(matplotlib.__version__)
```

**Matplotlib Pyplot**

---



submodule: **pyplot** and imported under the **plt** alias.

```python
# draw a line in a diagram from position (0, 0) to position (6, 250)

import matplotlib.pyplot as plt
import numpy as np

x_points=np.array([0, 6])
y_points=np.array([0, 250])

plt.plot(x_points, y_points)
plt.show()
```

**Matplotlib Plotting**

---



The **plot()** function is used to draw points in a diagram.

The funciton takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x axis.

Parameter 2 is an array containing the points on the y axis.

```python
# draw a line in a diagram from position (1, 3) to position (8, 10)

import matplotlib.pyplot as plt
import numpy as np

x_points=np.array([1, 8])
y_points=np.array([3, 10])

plt.plot(x_points, y_points)
plt.show()
```

**Plotting Without Line**

```python
# draw two points in the diagram, one at position (1, 3) and one in position (8, 10)

import matplotlib.pyplot as plt
import numpy as np

x_points=np.array([1, 8])
y_points=np.array([3, 10])

# 'o' means rings that is used to shortcut string natation parameter
plt.plot(x_points, y_points, 'o')
plt.show()
```

**Multiple Points**

```python
# draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and (8, 10)

import matplotlib.pyplot as plt
import numpy as np

x_points=np.array([1, 2, 6, 8])
y_points=np.array([3, 8, 1, 10])

plt.plot(x_points, y_points)
plt.show()
```

```python
# plotting without line from position (1, 3) to (2, 8) then to (6, 1) and (8, 10)

import matplotlib.pyplot as plt
import numpy as np

x_points=np.array([1, 2, 6, 8])
y_points=np.array([3, 8, 1, 10])

plt.plot(x_points, y_points, 'o')
plt.show()
```

**Default X-Points**

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3, 4, ....

```python
# plotting without x-points

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points)
plt.show()
```

**Matplotlib Markers**

---



We use the keyword argument **marker** to emphasize each point with a specified marker.

```python
# mark each point with a circle

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points, marker='o')
plt.show()
```

```python
# mark each point with a '*'

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points, marker='*')
plt.show()
```

**Format Strings fmt**

Shortcut string notation parameter is **fmt**

**Color reference:**

for Red 'r'

for Green 'g'

for Blue 'b'

for Cyan 'c'

for Magenta 'm'

for Yellow 'y'

for Black 'k'

for White 'w'

```python
# mark each point with a circle

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points, 'o:r')
plt.show()
```

```python
# mark each point with solid line

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points, 'o-k')
plt.show()
```

```python
# mark each point with dotted line

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points, 'o:')
plt.show()
```

```python
# mark each point with dashed line

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points, 'o--')
plt.show()
```

```python
# mark each point with dashed or dotted line

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points, 'o-.g')
plt.show()
```

**Marker Size**

```python
# set the size of the markers to 20

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

# ms means markersize
plt.plot(y_points, 'o-.g', ms=20)
plt.show()
```

**Marker Color**

```python
# set the edge color to blue

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

# mec means markeredgecolor
plt.plot(y_points, 'o-.y', ms=20, mec='b')
plt.show()
```

```python
# set the face color to white

import matplotlib.pyplot as plt
import numpy as np

y_points =np.array([3, 8, 1, 10, 5, 7])

# mfc means markerfacecolor
plt.plot(y_points, 'o-.y', ms=20, mec='b', mfc='w')
plt.show()
```
